```python
dataframe=pd.read_csv("desktop/Cars93.txt")
display(dataframe)
dataframe.info()
dataframe.describe()
object_columns = dataframe.select_dtypes('object').columns

number_columns = dataframe.select_dtypes(exclude='object').columns
```

2. For each numerical column, draw a histogram using pandas library.

```python
dataframe.hist(figsize=(15,20),color="r")
plt.show()
```

3. For each numerical column, draw a histogram using seaborn library.

```python
import seaborn as sns
object_columns = dataframe.select_dtypes('object').columns

number_columns = dataframe.select_dtypes(exclude='object').columns

for c in dataframe[number_columns].columns:
    plt.figure() 
    sns.histplot(x=c,data=dataframe,bins=10)
```

4. For each non-numerical column except *Make* and *Model*, draw a countplot (Hint: Use drop() function).

```python
object_columns = dataframe.select_dtypes('object').columns
newdataframe = object_columns.drop(['Make', 'Model'])

for i in dataframe[newdataframe].columns:
    plt.figure()
    sns.countplot(x=i,data=dataframe);
```

5. For each *Manufacturer*, draw countplot differentiated by *DriveTrain*

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.countplot(y='Manufacturer',data=dataframe, hue="DriveTrain");
```

6. Draw box-plots for the *Price, MPG.City, MPG.Highway, Cylinders, EngineSize,Turn.circle*  and differentiate by *Origin* column.

```python
for i in dataframe[number_columns].columns:
    plt.figure()
    sns.boxplot(data=dataframe, x="Origin",y=i)
```

7. For Cylinders column, find the row which has non-numerical entry and replace it with 0. Convert the data type of Cylinders into integer. (use lambda).(Hint: Try sorting to find the non-numerical entry)

```python
test=lambda x : True if (x.isnumeric()==True) else False
for i in dataframe.index:
    if(test(dataframe.loc[i,"Cylinders"])):
        display(print())
    else:
        dataframe.at[i,"Cylinders"]=0
dataframe.sort_values(by="Cylinders")        
```

8. Draw scatterplots between *Cylinders* and *EngineSize* and *Cylinders* and *horsepower*.

```python
sns.scatterplot(x="Cylinders",y="EngineSize",data=dataframe,color='red')
plt.figure()
sns.scatterplot(x="Cylinders",y="Horsepower",data=dataframe, color='blue')
```

9. Draw scatterplot between Cylinders and EngineSize, where the size of the marker is proportioned to horsepower (Use range of sizes from (10,100)).

```python
sns.scatterplot(x="Cylinders",y="EngineSize",data=dataframe, sizes=(10,100), size='Cylinders')
```
