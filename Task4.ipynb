5. Print all the rows for which city mileage is 20 or more and highway mileage is 30 or more and the car is front wheel drive. How many models are there?

```python
read_dataFrame= pd.read_csv('Cars93.txt')
store = ((read_dataFrame['MPG.highway'] >= 20) | (read_dataFrame['MPG.city'] >= 30)) & (read_dataFrame['DriveTrain'] == 'Front')
read_dataFrame.loc[store]
display(read_dataFrame.Model.count())
```

11. For the column "Man.trans.avail", using lambda() function, convert it to 0 for Yes and 1 for No.

```python
read_dataFrame= pd.read_csv('Cars93.txt')
read_dataFrame['Man.trans.avail']=read_dataFrame['Man.trans.avail'].apply(lambda x: x.replace('0','Yes'),('1','No'))
display(read_dataFrame.head())
```
