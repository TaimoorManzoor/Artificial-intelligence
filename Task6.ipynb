```python
#import statements
#required libraries
import requests
from bs4 import BeautifulSoup
```

1. Request the content from the website, 'https://homeshopping.pk/'.

```python
page = requests.get('https://homeshopping.pk/')
page
```

2. Display the status code, was the request successful?

```python
if page.status_code==200:
    display(page.status_code)
    print('Pass')
else:
    display(page.status_code)
    print('Fail')
```

3. Parse the contents of the requested website using BeautifulSoup library, use the 'html.parser'. Print it with indentation


```python
soup = BeautifulSoup(page.content, 'html.parser')
display(soup)
```

4. For one product, gather the following:
    - product name
    - product ID
    - product price
    - link to the product page
    
Remove empty spaces where ever necessary.

Start by looking at the website and identify relevant tags.

```python
info = soup.find('div', class_='product-box')
print(info.find('span').get('data-product_id'))
print(info.find(class_='ProductDetails').get_text())
print(info.find(class_='ActualPrice').get_text())
print(info.find('a').get('href'))
```

5. Perform step 4 for all the products, using `for` loop, store the data as list.

```python
list_product=[]
for info in soup.find_all('div', class_='product-box'):
    o=[info.find('span').get('data-product_id'),info.find(class_='ProductDetails').get_text(),info.find(class_='ActualPrice').get_text(),info.find('a').get('href')]
    list_product.append(o)
list_product

    
```

6. Convert the list (from step 5) into dataframe object, give each column an appropriate name.

```python
import pandas as pd

dataframe = pd.DataFrame(list_product,columns=['Product ID','Product name','Product Prices','Links'])
display(dataframe)
```
